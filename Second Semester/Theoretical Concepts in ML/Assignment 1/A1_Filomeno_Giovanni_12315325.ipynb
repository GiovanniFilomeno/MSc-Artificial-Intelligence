## Exercise 1: Bayes' Theorem

Prove Bayes' theorem that says
\begin{equation*}
    p(X \mid Y) = \frac{p(Y \mid X) p(X)}{p(Y)}.
\end{equation*}

########## YOUR SOLUTION HERE ##########

The conditional probability of $X$ given $Y$ is defined as:

$$
p(X \mid Y) = \frac{p(X \cap Y)}{p(Y)},
$$

where $p(X \cap Y)$ is the probability of both $X$ and $Y$ happening, and $p(Y)$ is the probability of $Y$ happening.

Similarly, the conditional probability of $Y$ given $X$ is:

$$
p(Y \mid X) = \frac{p(X \cap Y)}{p(X)},
$$

where $p(X)$ is the probability of \(X\) happening.

From the second equation, we can express $p(X \cap Y)$ as:

$$
p(X \cap Y) = p(Y \mid X) \cdot p(X).
$$

Substituting this expression for $p(X \cap Y)$ into the first equation gives the Bayes' theorem:

$$
p(X \mid Y) = \frac{p(Y \mid X) \cdot p(X)}{p(Y)}.
$$

## Exercise 2: Expected Value of the Uniform Distribution

Let $X$ be a random variable drawn from the continuous uniform distribution on the interval $[a, b) \subset \mathbb R$. 
Prove that the expected value $\mathbb E[X] = \frac{a+b}{2}$. 

########## YOUR SOLUTION HERE ##########

For a continuous uniform distribution on $[a,b)$, the PDF is defined as:

$$
p(X) = \left\{\begin{matrix}
\frac{1}{b-a} & \text{for } a \leq x < b, \\ 
0 & \text{otherwise}
\end{matrix}\right.
$$

The expected value of $X$ is defined as:

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) dx
$$

Combining the two equation the following expression can be obtained:

$$
\mathbb{E}[X] = \int_{a}^{b} x \frac{1}{b-a} dx = \frac{b^2}{2(b-a)} - \frac{a^2}{2(b-a)}  = \frac{(b+a)(b-a)}{2(b-a)} = \frac{(b+a)}{2}
$$

## Exercise 3: Variance of the Uniform Distribution

Let $X$ be a random variable drawn from the continuous uniform distribution on the interval $[a, b) \subset \mathbb R$. 
Prove that $ \operatorname{Var}(X) = \frac{(a-b)^2}{12}$. 
What are the interval boundaries of a centered continuous uniform random variable with unit variance?

########## YOUR SOLUTION HERE ##########

To compleate this exercise we can use the definition of variance $\operatorname{Var}(X)$ of a random variable $X$:

$$
\operatorname{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

The second part, $(\mathbb{E}[X])^2$ can be taken from the previous exercises and it is equal to $\frac{(b+a)^2}{4}$.
The second part can be calculated using the previous definiton of expected value using this calculation: 

$$
\mathbb{E}[X^2] = \int_{a}^{b} x^2 \frac{1}{b-a} dx = \frac{b^3-a^3}{3(b-a)} = \frac{(b-a)^3+3b^2 -3a^2 b}{3(b-a)}=\frac{(b-a)^2}{3}+ab
$$

Combinining the two expected values: 

$$
\operatorname{Var}(X)=\frac{(b-a)^2}{3}+ab - \frac{(b+a)^2}{4} = \frac{4(b-a)^2+12ab-3(b+a)^2}{12} = \frac{a^2+b^2-2ab}{12}=\frac{(a-b)^2}{12}
$$

For the uniform distribution to be centered at $0$, we need that the expected value $\frac{(b+a)}{2}=0$ while the variance $\frac{(a-b)^2}{12} = 1$.
From the expected value we obtain that $b=-a$. Substituting into the variance imposed to 1, we obtain that $b=\sqrt{3}$ while $a=-\sqrt{3}$.
Therefore, the interval boundaries of a centered continuous uniform random variable with unit variance are $[- \sqrt{3}, \sqrt{3})$.

## Exercise 4: CLT Simulation

Write a python routine that demonstrates the central limit theorem (CLT). 
Draw $n$ samples $X_1, \dots, X_n$ from the uniform distribution over $[-\sqrt{3},\sqrt{3})$. 
Then compute $Z = \frac{1}{\sqrt{n}} \sum_{i=1}^n X_i$. 
Compute 100000 realizations of $Z$ and summarize them in a density or histogram plot. 
Compare the resulting curves to the density of the standard normal distribution. 
Perform this procedure for $n \in \{1, 2, 10\}$. 
What do you observe and how do your observations relate to the CLT?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy.stats import norm

%matplotlib inline

########## YOUR SOLUTION HERE ##########

num_realizations = 100000
n_values = [1, 2, 10]
a, b = -np.sqrt(3), np.sqrt(3)

plt.figure(figsize=(18, 6))

for i, n in enumerate(n_values, 1):
    samples = np.random.uniform(low=a, high=b, size=(num_realizations, n))
    Z = np.sum(samples, axis=1) / np.sqrt(n)
    # Z = np.mean(samples, axis = 1) * np.sqrt(n)
    
    plt.subplot(1, len(n_values), i)
    sns.histplot(Z, bins=50, kde=True, stat='density', label=f'n = {n}')
    plt.title(f'Distribution of Z for n = {n}')
    plt.xlabel('Z')
    plt.ylabel('Density')
    
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, norm.pdf(x), 'r--', label='Standard Normal')
    plt.legend()

plt.tight_layout()
plt.show()
```

The graphs shows that increasing the size of $n$, the distribution of the sample mean tends to be a normal distribution
