The cumulant-generating function $K_X(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
  \newcommand\E{\operatorname{\mathbb{E}}}
  \newcommand\dif{\mathrm{d}}
  K_X(t) = \log \E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n(X) = K_X^{(n)}(0)$ the $n$-th derivative of $K_X(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
  K_X(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1 (20 Points)</h2>
Show that if the random variables $X$ and $Y$ are independent then $\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds. 

########## YOUR SOLUTION HERE ##########

##### Part 1: $\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ for independent $X$ and $Y$

Since $X$ and $Y$ are independent, their moment-generating functions multiply:
$$
E[e^{t(X+Y)}] = E[e^{tX}]E[e^{tY}].
$$

Thus, the cumulant-generating function for $X+Y$ is:
$$
K_{X+Y}(t) = \log E[e^{t(X+Y)}] = \log (E[e^{tX}]E[e^{tY}]) = \log E[e^{tX}] + \log E[e^{tY}] = K_X(t) + K_Y(t).
$$

Taking derivatives and evaluating at $t = 0$ and using the properties of derivatives,
$$
\frac{\partial^n}{\partial t^n} K_{X+Y}(t) \bigg|_{t=0} = \frac{\partial^n}{\partial t^n} (K_X(t) + K_Y(t)) \bigg|_{t=0} = \frac{\partial^n}{\partial t^n} K_X(t) \bigg|_{t=0} + \frac{\partial^n}{\partial t^n} K_Y(t) \bigg|_{t=0}.
$$

Given that $\kappa_n(X) = K_X^{(n)}(0)$ and $\kappa_n(Y) = K_Y^{(n)}(0)$, it follows that
$$
\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y).
$$

##### Part 2: $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$

Given the scaling property of exponential functions,
$$
E[e^{t(\alpha X)}] = E[e^{(\alpha t)X}].
$$

Thus, the cumulant-generating function for $\alpha X$ is:
$$
K_{\alpha X}(t) = \log E[e^{(\alpha t)X}].
$$

When taking the $n$-th derivative of $K_{\alpha X}(t)$, apply the chain rule:
$$
\frac{\partial^n}{\partial t^n} K_{\alpha X}(t) = \frac{\partial^n}{\partial t^n} \log E[e^{(\alpha t)X}] = \frac{\partial^n}{\partial (\alpha t)^n} \log E[e^{(\alpha t)X}] \cdot (\alpha^n),
$$
where $\alpha^n$ arises from differentiating $(\alpha t)^n$.

Evaluating at $t=0$ gives:
$$
K_{\alpha X}^{(n)}(0) = \alpha^n K_X^{(n)}(0),
$$

So, it follows that:
$$
\kappa_n(\alpha X) = \alpha^n \kappa_n(X).
$$

<h2 style="color:rgb(0,120,170)">Task 2 (20 Points)</h2>

Show that the first four cumulants of a centered random variable $X$, i.e. $E[X] = 0$, are
\begin{equation}
  \kappa_1 = 0,\quad\kappa_2 = E[X^2],\quad\kappa_3 = E[X^3],\quad\text{and}\quad\kappa_4 = E[X^4] - 3(E[X^2])^2,
\end{equation}
respectively. *Hint:* Combine the Taylor expansions $E[e^{tX}] = \sum_{n=0}^\infty E[X^n]\frac{t^n}{n!}$ and $\log(1 + a) = \sum_{n=1}^\infty a^n \frac{(-1)^{n+1}}{n}$ and drop all terms of order higher than 4. 

########## YOUR SOLUTION HERE ##########

For the centered random variable $X$, the MGF can be expanded as:
$$
E[e^{tX}] = 1 + E[X]\frac{t}{1!} + E[X^2]\frac{t^2}{2!} + E[X^3]\frac{t^3}{3!} + E[X^4]\frac{t^4}{4!} + \cdots
$$

Since $E[X] = 0$ for a centered random variable, this simplifies to:
$$
E[e^{tX}] = 1 + E[X^2]\frac{t^2}{2!} + E[X^3]\frac{t^3}{3!} + E[X^4]\frac{t^4}{4!} + O(t^5)
$$

The cumulant-generating function $K_X(t)$ is the natural logarithm of the MGF:
$$
K_X(t) = \log E[e^{tX}]
$$

Usingse the Taylor expansion for $\log(1 + a)$:
$$
\log(1 + a) = \sum_{n=1}^\infty (-1)^{n+1}\frac{a^n}{n}
$$

Using $a = E[e^{tX}] - 1$:
$$
K_X(t) = \log \left(1 + E[X^2]\frac{t^2}{2!} + E[X^3]\frac{t^3}{3!} + E[X^4]\frac{t^4}{4!} + O(t^5)\right)
$$

This expands (dropping terms of order higher than 4 and using $ \frac{t^2}{2!} = \frac{t^2}{2} and \frac{t^3}{3!} = \frac{t^3}{6} and \frac{t^4}{4!} = \frac{t^4}{24} $):
$$
K_X(t) = \left( E[X^2]\frac{t^2}{2} + E[X^3]\frac{t^3}{6} + E[X^4]\frac{t^4}{24} \right) - \frac{1}{2}\left(E[X^2]\frac{t^2}{2}\right)^2 + O(t^5)
$$

Simplify and collect like terms:
$$
K_X(t) = 0\cdot t + E[X^2]\frac{t^2}{2} + E[X^3]\frac{t^3}{6} + \left(E[X^4]\frac{t^4}{24} - \frac{1}{2}\left(\frac{E[X^2]t^2}{2}\right)^2\right) + O(t^5)
$$

By comparing this to the series expansion $K_X(t) = \sum_{n=1}^\infty \kappa_n \frac{t^n}{n!}$:
- $\kappa_1 = 0$ (no $t$ term)
- $\kappa_2 = E[X^2]$
- $\kappa_3 = E[X^3]$
- $\kappa_4 = E[X^4] - 3E[X^2]^2$ (as $\frac{1}{2}(\frac{E[X^2]^2t^4}{4}) = 3E[X^2]^2\frac{t^4}{24}$)

<h2 style="color:rgb(0,120,170)">Task 3 (20 Points)</h2>

Consider a standard normal random variable $X \sim p(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$. Show that $E[X] = 0$ and $E[X^n] = (n-1) E[X^{n-2}]$ for $n \geq 2$ and use this to determine the third and fourth cumulants of $X$. *Hint:* Use the fact that $\int f'(x) e^{f(x)} \partial x = e^{f(x)}$. 

########## YOUR SOLUTION HERE ##########

For a standard normal random variable $X$, its PDF is symmetric around zero. Therefore $E[X]$ for a symmetric distribution centered at zero is:
$$
E[X] = \int_{-\infty}^\infty x p(x) \, dx = \int_{-\infty}^\infty x \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \, dx
$$
Due to the symmetry, and due to the fact that the function $x e^{-\frac{x^2}{2}}$ is an odd function, the integral over the symmetric limits results in zero ($E[X] = 0$).

##### Second task
Considering this integral:
$$
\int x^n e^{-\frac{x^2}{2}} \, dx
$$
can be rewritten by recognizing that differentiation of $x^{n-1}$ leads to a reduction by two orders upon integration by parts, and also by using the hint. The differentiation of the exponential part $e^{-\frac{x^2}{2}}$ yields $-x e^{-\frac{x^2}{2}}$.

Specifically, integrating by parts with $u = x^{n-1}$ and $dv = x e^{-\frac{x^2}{2}} \, dx$, using the derivative $du = (n-1)x^{n-2} \, dx$ and the integral $v = -e^{-\frac{x^2}{2}}$:
$$
\int x^n e^{-\frac{x^2}{2}} \, dx = x^{n-1}(-e^{-\frac{x^2}{2}}) \Big|_{-\infty}^\infty - \int_{-\infty}^\infty (n-1)x^{n-2}(-e^{-\frac{x^2}{2}}) \, dx
$$

The boundary terms goes to zero:
$$
\int x^n e^{-\frac{x^2}{2}} \, dx = (n-1) \int x^{n-2} e^{-\frac{x^2}{2}} \, dx
$$
This proves the recurrence relation:
$$
E[X^n] = (n-1)E[X^{n-2}]
$$

From the recurrence relation, since $E[X^3] = 2E[X^1]$ and $E[X^1] = 0$, we have $E[X^3] = 0$. Thus, $\kappa_3 = E[X^3] = 0$.

For $E[X^4]$:
$$
E[X^4] = 3E[X^2]
$$

Since  $E[X^2] = 1$ (by the definition of a standard normal random variable):
$$
E[X^4] = 3
$$

Hence, $\kappa_4 = E[X^4] - 3(E[X^2])^2 = 3 - 3(1)^2 = 0$.

<h2 style="color:rgb(0,120,170)">Task 4 (20 Points)</h2>

Compare PCA and ICA at the example of the `cifar-10.npy` dataset. Load the dataset, perform PCA and ICA using the implementations of the `sklearn.decomposition` package with 100 components each. Then visualize the components extracted by the two methods using the provided functions `show_image` and `scale`. What differences do you observe and how do they relate to the properties of PCA and ICA? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA

def scale(s, a=0, b=1):
    return (b - a) * (s - s.min()) / (s.max() - s.min()) + a

def show_image(img):
    plt.imshow(img.reshape(3, 32, 32).transpose((1, 2, 0)))
```

```python
########## YOUR SOLUTION HERE ##########
data = np.load('cifar-10.npy') 

# Perform PCA
pca = PCA(n_components=100)
pca_components = pca.fit_transform(data)

# Perform ICA
ica = FastICA(n_components=100)
ica_components = ica.fit_transform(data)

# Original Images
pca_images = pca.inverse_transform(pca_components)
ica_images = ica.inverse_transform(ica_components)

def visualize_components(components, title):
    plt.figure(figsize=(10, 10))
    for i in range(10):
        plt.subplot(5, 5, i + 1)
        show_image(scale(components[i]))
        plt.xticks([])
        plt.yticks([])
    plt.suptitle(title)
    plt.show()

# Visualize PCA and ICA components
visualize_components(pca_images, 'PCA Components')
visualize_components(ica_images, 'ICA Components')
```

Consider the cocktail party problem. Read the left channel of the files `birds.wav`, `ocean.wav`, `frogs.wav` using the `scipy.io.wavfile` package. Then construct a full-rank $3 \times 3$ mixing matrix and produce 3 different mixings of those signals. You can listen to them using the `sounddevice` package. Finally, use `FastICA` from the `sklearn.decomposition` package to recover the original sources. Listen to the recovered sources and compare them to the originals to check if ICA was successful. *Hint:* Scale the signals to $[-1,1]$ before playing them. 

```python
import scipy.io.wavfile
import sounddevice as sd

########## YOUR SOLUTION HERE ##########

from sklearn.decomposition import FastICA

# Read audio files --> taking the sampling for birds for playing the sound later
rate_birds, birds = scipy.io.wavfile.read('birds.wav')
_, ocean = scipy.io.wavfile.read('ocean.wav')
_, frogs = scipy.io.wavfile.read('frogs.wav')

birds = birds[:, 0] if birds.ndim == 2 else birds
ocean = ocean[:, 0] if ocean.ndim == 2 else ocean
frogs = frogs[:, 0] if frogs.ndim == 2 else frogs

# Create a full rank 3x3 mixing matrix
mixing_matrix = np.random.rand(3, 3)

# Make sure it is full rank
assert np.linalg.matrix_rank(mixing_matrix) == 3

# Mix the signals
mixed_signals = np.dot(np.c_[birds, ocean, frogs], mixing_matrix)

def scale_signal(signal):
    return 2. * (signal - np.min(signal)) / np.ptp(signal) - 1

# Scale the mixed signals to [-1, 1]
mixed_signals_scaled = scale_signal(mixed_signals)

# Play the mixed signals
for i in range(mixed_signals_scaled.shape[1]):
    print(f"Playing mixed signal {i+1}")
    sd.play(mixed_signals_scaled[:, i], rate_birds)
    sd.wait()

ica = FastICA(n_components=3)

recovered_signals = ica.fit_transform(mixed_signals_scaled)
recovered_signals_scaled = scale_signal(recovered_signals)

# Play the recovered signals
for i in range(recovered_signals_scaled.shape[1]):
    print(f"Playing recovered signal {i+1}")
    sd.play(recovered_signals_scaled[:, i], rate_birds)
    sd.wait()
```
