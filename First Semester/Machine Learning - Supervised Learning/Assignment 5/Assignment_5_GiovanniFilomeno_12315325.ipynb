<h2 style="color:rgb(0,120,170)">Task 1: AdaBoostM1 is an instance of forward stagewise modelling</h2>


In the lecture it was mentioned that one of the first boosting algorithms, i.e. AdaBoostM1, is equivalent to forward stagewise modelling using the exponential loss $L(y,g(\mathbf{x}))=\exp(-yg(\mathbf{x}))$ for a binary classification problem ( $y\in \left\{ -1,1 \right\}$ ). In this task we intend to provide proof of this fact. We will guide you through the most important steps and you will have to add some details.

For AdaBoostM1, the basis functions at timestep $n$ are the individual classifiers $b_n(\mathbf{x}) \in \left\{ -1,1 \right\}$. We assume that all of them are slightly better than random guessing. Note that we use $b_n$ here for the resulting classifier at timestep $n$, which differs slightly from the notation in the slides, mainly to not confuse it with the corresponding approximation from forward stagewise modelling, which is also called $g_n$ there. 

Using the exponential loss in each timestep $n$ we have to solve 
$$
(\beta_n,b_n)= \arg \min_{\beta,b} \sum_{i=1}^l \exp(-y_i(g_{n-1}(\mathbf{x}_i)+\beta b(\mathbf{x}_i)))
$$ 

for the classifier $b_n$ and the coefficient $\beta_n$ which are added at each step. This can be rewritten as
$$
(\beta_n,b_n)= \arg \min_{\beta,b} \sum_{i=1}^l w_i^{(n)} \exp(-y_i \beta b(\mathbf{x}_i)) \qquad (1)
$$ 

with $w_i^{(n)}=\exp(-y_i g_{n-1}(\mathbf{x}_i))$. Since each $w^{(n)}_i$ depends neither on $\beta$ nor $b$, it can be regarded as a weight that is applied to each observation. This weight depends on $g_{n−1}(\mathbf{x}_i)$, and so the individual weight values change with each iteration $n$. The solution of (1) can be found in two steps:

$$
\left ( \beta_n, b_n \right ) = \arg \min_{\beta, b} \sum_{i=1}^{l} w_i^{(n)} \exp\left ( -y_i \beta b (\mathbf{x}_i) \right )

Rewriting it and separating the summation into two parts: when $b(\mathbf{x}_i) = y_i$ and when $b(\mathbf{x}_i) \ne y_i$.

When $b(\mathbf{x}_i) = y_i$, the expression $\exp(-y_i \beta b(\mathbf{x}_i))$ becomes $\exp(-\beta)$, since $y_i b(\mathbf{x}_i) = 1$.
When $b(\mathbf{x}_i) \ne y_i$, the expression becomes $\exp(\beta)$, since $y_i b(\mathbf{x}_i) = -1$.
So the rewritten expression is:

$$
\sum_{b(\mathbf{x}_i)=y_i} w_i^{(n)} \exp(-\beta) + \sum_{b(\mathbf{x}_i) \neq y_i} w_i^{(n)} \exp(\beta)

$$ \exp(\beta) \sum_{b(\mathbf{x}_i) \ne y_i} w_i^{(n)} + \exp(-\beta) \sum_{b(\mathbf{x}_i) = y_i} w_i^{(n)} + \exp(-\beta) \sum_{b(\mathbf{x}_i) \ne y_i} w_i^{(n)} $$

- Hint 3

The aim is to minimize this simplified expression. Considering that the term $\exp(-\beta) \sum_{i=1}^l w_i^{(n)}$ is constant with respect to the classifier $b$. Therefore, the minimization depends only on the first term: $(\exp(\beta) - \exp(-\beta)) \sum_{b(\mathbf{x}_i) \ne y_i} w_i^{(n)}$.

Since $\beta \geq 0$, $\exp(\beta) - \exp(-\beta)$ is always positive. Therefore, to minimize the entire expression, is necessary to minimize $\sum_{b(\mathbf{x}_i) \ne y_i} w_i^{(n)}$, which is the sum of weights where the classifier $b$ disagrees with the actual label $y_i$.

This leads to the conclusion that the solution to (1) is indeed:

$$
b_n = \arg \min_b \sum_{i=1}^{l} w_i^{(n)} I(b(\mathbf{x}_i) \ne y_i)

In the **second step** you need to optimize the following expression with respect to $\beta$:$$(\exp(\beta)-\exp(-\beta))\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)}+ \exp(-\beta) \sum_{i=1}^l w_i^{(n)}$$ Hint: do it in the usual way (i.e. by differentiating the expression and setting it to $0$). <br><br>

1) Derivative with respect to $\beta$. The derivative of $\exp(\beta)$ is again $\exp(\beta)$, while the derivative of $\exp(-\beta)$ is $-\exp(-\beta)$:

$$
\frac{d}{d\beta} \left[ (\exp(\beta) - \exp(-\beta))\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)} + \exp(-\beta) \sum_{i=1}^l w_i^{(n)} \right] = \left( \exp(\beta) + \exp(-\beta) \right)\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)} - \exp(-\beta) \sum_{i=1}^l w_i^{(n)}
$$

2) Imposing the derivative equal to zero to find the optimum

$$
\left( \exp(\beta) + \exp(-\beta) \right)\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)} - \exp(-\beta) \sum_{i=1}^l w_i^{(n)} = 0
$$

3) Rearranging the formula and isolating $\beta$:

Rearranging leads to:

$$
\exp(\beta) \sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)} = \exp(-\beta) \left( \sum_{i=1}^l w_i^{(n)} - \sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)} \right)
$$

Taking the natural logarithm to isolate to isolate the $\beta$, and considering that $\exp(\beta) / \exp(-\beta) = \exp(2 \beta)$

$$
\beta = \frac{1}{2} \ln \frac{\sum_{i=1}^l w_i^{(n)} - \sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)}}{\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)}}
$$

4) Simplify
Dividing the numerator and denominator inside the natural logaritmic by $\sum_{i=1}^l w_i^{(n)}$ and using the abbreviation $\text{err}_n = \frac{\sum_{i=1}^l w_i^{(n)} I(y_i \ne b_n(\mathbf{x}_i))}{\sum_{i=1}^l w_i^{(n)}} = \frac{\sum_{b_n(\mathbf{x}_i)\ne y_i}w_i^{(n)}}{\sum_{i=1}^l w_i^{(n)}}$ the formula leads to: 

$$

$$


<h3 style="color:rgb(210,90,80)">Calculation 3 (10 points):</h3>

In the **final step** we can update the approximation as follows: $g_n(\mathbf{x})=g_{n-1}(\mathbf{x})+\beta_n b_n(\mathbf{x})$. To finish the proof proceed by deriving the following relations:
1) The weights for the next generation can be computed as follows: $w_i^{(n+1)}=w_i^{(n)} \exp(-y_i \beta_n b_n(\mathbf{x}))$.
2) $-y_i b_n(\mathbf{x})=2I(y_i \ne b_n(\mathbf{x}_i))-1$
3) Use these two relations to show that: $$w_i^{(n+1)}=w_i^{(n)} \exp(-\beta_n) \exp(\alpha_n I(y_i \ne b_n(\mathbf{x}_i))) \qquad (2)$$ where $\alpha_n=2\beta_n$ is the $\alpha_n$ from the AdaBoostM1 algorithm from the lecture.

**Please provide reasoning and explanations in full sentences. Grading of the task will heavily depend on it.**

1 - Calculating Weights for the Next Generation

Given the update rule for the approximation $g_n(\mathbf{x}) = g_{n-1}(\mathbf{x}) + \beta_n b_n(\mathbf{x})$, it is necessary to calculate the weights for the next iteration $w_i^{(n+1)}$.

From the definition that $w_i^{(n+1)} = \exp(-y_i g_n(\mathbf{x}_i))$. Substituting the update rule into this definition:

$$
w_i^{(n+1)} = \exp(-y_i (g_{n-1}(\mathbf{x}_i) + \beta_n b_n(\mathbf{x}_i)))
$$

Expanding this, using the fact that $w_i^{(n)} = \exp(-y_i g_{n-1}(\mathbf{x}_i))$:

$$
w_i^{(n+1)} = w_i^{(n)} \exp(-y_i \beta_n b_n(\mathbf{x}_i))
$$

2 - Relation between Classifier Output and Indicator Function

The next step is to show the relation $-y_i b_n(\mathbf{x}) = 2I(y_i \ne b_n(\mathbf{x}_i)) - 1$.

- If $y_i = b_n(\mathbf{x}_i)$, then $-y_i b_n(\mathbf{x}_i) = -1$ and $I(y_i \ne b_n(\mathbf{x}_i)) = 0$. So, $2 \times 0 - 1 = -1$.
- If $y_i \ne b_n(\mathbf{x}_i)$, then $-y_i b_n(\mathbf{x}_i) = 1$ and $I(y_i \ne b_n(\mathbf{x}_i)) = 1$. So, $2 \times 1 - 1 = 1$.

Thus, the relation holds.

3 - Showing the Final Relation for Weight Update

Using the two relations derived above, $w_i^{(n+1)}$ can be expressed as follows:

Starting with the expression from (1):
$$
w_i^{(n+1)} = w_i^{(n)} \exp(-y_i \beta_n b_n(\mathbf{x}_i))
$$

Substituting the relation from (2) into this:
$$
w_i^{(n+1)} = w_i^{(n)} \exp((2I(y_i \ne b_n(\mathbf{x}_i)) - 1) \beta_n)
$$

This can be rewritten as:
$$
w_i^{(n+1)} = w_i^{(n)} \exp(-\beta_n) \exp(2 \beta_n I(y_i \ne b_n(\mathbf{x}_i)))
$$

Since  $\alpha_n = 2\beta_n$, the final expression becomes:
$$
w_i^{(n+1)} = w_i^{(n)} \exp(-\beta_n) \exp(\alpha_n I(y_i \ne b_n(\mathbf{x}_i)))
$$

```python
def _filter_(
        x_train: np.ndarray, 
        y_train: np.ndarray, 
        x_test: np.ndarray, 
        y_test: np.ndarray, 
        labels_list: list[int, int]
) -> tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
    """This function filters the datasets w.r.t to given labels_list.
    So, in the end x_train and x_test only contain the samples with labels that are in labels_list and y_train and y_test only contain 2 labels (i.e. 0 and 1).

    Parameters
    ----------
    x_train : np.ndarray
        Training data matrix.
    y_train : np.ndarray
        Training labels vector.
    x_test : np.ndarray
        Test data matrix.
    y_test : np.ndarray
        Test labels vector.
    labels_list : list[int, int]
        list of length 2 which consists of integer labels.
        
    Returns
    -------
    tuple(np.ndarray, np.ndarray, np.ndarray, np.ndarray)
        Returns the filtered train and test data matrix and labels vector (which only conists of 0s and 1s now).
        Check the return statement to see the actual order of returned arrays.
    """
    #Your code goes here ↓↓↓
    
    x_train_filtered = None
    y_train_filtered = None
    
    x_test_filtered = None
    y_test_filtered = None
    #Your code ends here _____________________________________________________________________________________

    # Filter training data
    train_mask = np.isin(y_train, labels_list)
    x_train_filtered = x_train[train_mask]
    y_train_filtered = np.where(y_train[train_mask] == labels_list[0], 1, 0)

    # Filter test data
    test_mask = np.isin(y_test, labels_list)
    x_test_filtered = x_test[test_mask]
    y_test_filtered = np.where(y_test[test_mask] == labels_list[0], 1, 0)
    
    return x_train_filtered, y_train_filtered, x_test_filtered, y_test_filtered
```

```python
def fit_predict(
        x_train: np.ndarray, 
        y_train: np.ndarray, 
        x_test: np.ndarray, 
        y_test: np.ndarray, 
        rseed: int
) -> tuple[RandomForestClassifier, np.ndarray]:
    """Function fits a RandomForestClassifier on the training data and returns predictions as well as model.

    Parameters
    ----------
    x_train : np.ndarray
        Training data matrix.
    y_train : np.ndarray
        Training labels vector.
    x_test : np.ndarray
        Test data matrix.
    y_test : np.ndarray
        Test labels vector.
    rseed : int
        Random Seed used for initializing the Classifier.

    Returns
    -------
    tuple(RandomForestClassifier, np.ndarray)
        Where the classifier is the already trained classifier and the array is the predictions on x_test.
    """
    #Your code goes here ↓↓↓
    model = None
    
    prediction = None
    #Your code ends here _____________________________________________________________________________________

    # Initialize the Random Forest with the random seed
    model = RandomForestClassifier(random_state=rseed)

    # Fir the model to the training data
    model.fit(x_train, y_train)

    # Make prediction on the test data
    prediction = model.predict(x_test)
    
    return model, prediction
```

Now, within the function `get_n_items_wrong` compute and return the size of the test set and number of misclassified test samples.

```python
def get_n_items_wrong(
        x_test:
        np.ndarray, y_test:
        np.ndarray,
        prediction: np.ndarray
) -> tuple[int, int, np.ndarray, np.ndarray, np.ndarray]:
    """Calculates the size of the test set, number of misclassified samples, the samples that are misclassified,
    the actual labels that are misclassified in the prediction, and the labels that are wrong in the prediction.
    
    Parameters
    ----------
    x_test : np.ndarray
        Test data matrix.
    y_test : np.ndarray
        Test labels vector.
    prediction : np.ndarray
        The predicted labels vector from the model.
    
    Returns
    -------
    tuple(int, int, np.ndarray, np.ndarray, np.ndarray)
        Where the first two ints are the size_test and num_wrong and the arrays are the samples_wrong, labels_wrong and predictions_wrong.
    """
    # Your code goes here ↓↓↓
    size_test = None
    num_wrong = None
    samples_wrong = None
    labels_wrong = None
    predictions_wrong = None
    # Your code ends here _____________________________________________________________________________________

    # Size of the test set
    size_test = x_test.shape[0]

    # Indices where prediction and labels are different
    wrong_indices = np.where(prediction != y_test)[0]
    # Num of wrong prediction as length of indices
    num_wrong = len(wrong_indices)

    # Misclassified samples and labels
    samples_wrong = x_test[wrong_indices]
    labels_wrong = y_test[wrong_indices]
    predictions_wrong = prediction[wrong_indices]
    
    return size_test, num_wrong, samples_wrong, labels_wrong, predictions_wrong
```

```python
def plot_wrong_predictions(samples_wrong: np.ndarray, labels_wrong: np.ndarray, predictions_wrong: np.ndarray) -> plt.Figure:
    """Function creates a figure that shows misclassified samples.

    Parameters
    ----------
    samples_wrong : np.ndarray
        The samples that got misclassified by the classifier.
    labels_wrong : np.ndarray
        The actual true labels of the misclassified samples.
    predictions_wrong : np.ndarray
        The predicted labels of the misclassified samples.

    Returns
    -------
    plt.Figure
        A matplotlib.pyplot figure object (i.e. the misclassified samples).
    """
    # Your code goes here ↓↓↓
    fig = None
    # Your code ends here _____________________________________________________________________________________

    # Limit to 20 or fewer samples
    num_samples = min(len(samples_wrong), 20)

    # Create a figure with a grid layout, round the number of rows to deal with num_samples=1
    fig, axes = plt.subplots(nrows=(num_samples + 4) // 5, ncols=5, figsize=(15, 3 * ((num_samples + 4) // 5)))
    fig.suptitle("Misclassified Samples")

    axes = axes.flatten()

    for i in range(num_samples):
        # Reshape sample to 2D image 28x28
        img = samples_wrong[i].reshape(28, 28)

        # Plot image with title showing true and predicted label
        axes[i].imshow(img, cmap='gray')
        axes[i].set_title(f"True: {labels_wrong[i]}, Pred: {predictions_wrong[i]}")
        axes[i].axis('off')

    # Turn off axes for empty plots
    for j in range(num_samples, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()

    return fig
```

```python
#PLOTTING HEATMAPS
def plot_heatmaps(x_train: np.ndarray, y_train: np.ndarray, model: RandomForestClassifier) -> plt.Figure:
    """Create plots as described in the task description above.

    Parameters
    ----------
    x_train : np.ndarray
        Training data matrix.
    y_train : np.ndarray
        Training labels vector.
    model : RandomForestClassifier
        Your already trained classifier.

    Returns
    -------
    plt.Figure
        A matplotlib.pyplot figure object (i.e. the heatmaps).
    """
    #Your code goes here ↓↓↓
    
    #Use your freestlye plotting
    
    #Step 1 Split and calculate averages
    dress_av = None
    bag_av = None
    
    #Step 2 Subtract 
    diff = None
    
    #Step 3 Extract feature importance
    importances = None
    
    #Step 4 plot everything together
    fig = None
    #Your code ends here _____________________________________________________________________________________

    #Step 1 Split and calculate averages
    dresses = x_train[y_train == 1]
    bags = x_train[y_train == 0]

    dress_av = np.mean(dresses, axis=0).reshape(28, 28)
    bag_av = np.mean(bags, axis=0).reshape(28,28)

    #Step 2 Subtract
    diff = dress_av - bag_av

    #Step 3 Extract feature importance
    importances = model.feature_importances_.reshape(28,28)

    #Step 4 plot everything together
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

    sns.heatmap(dress_av, ax=axes[0], cmap='viridis')
    axes[0].set_title('Average Dress')

    sns.heatmap(bag_av, ax=axes[1], cmap='viridis')
    axes[1].set_title('Average Bag')

    sns.heatmap(diff, ax=axes[2], cmap='viridis')
    axes[2].set_title('Overlaps')

    plt.figure(figsize=(6, 6))
    sns.heatmap(importances, cmap='viridis')
    plt.title('Feature Importances')

    return fig
```
