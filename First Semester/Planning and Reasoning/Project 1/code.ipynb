```python
import subprocess

def run_limboole_with_exe(input_string, exe_path="limbooleAPE.exe"):

    try:
        # Esegui Limboole utilizzando il file .exe
        process = subprocess.run(
            [exe_path, "-s"],
            input=input_string,
            text=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE
        )
        # Restituisco l'output o l'errore
        if process.stderr:
            return f"Errore: {process.stderr}"
        else:
            return process.stdout
    except subprocess.SubprocessError as e:
        return f"Errore nell'esecuzione: {e}"
```

```python
def find_empty_tile(state):
    for i in range(len(state)):
        for j in range(len(state[i])):
            if state[i][j] == 0:
                return (i, j)
    return None


def create_movement_code(time, moving_tile, current_pos, new_pos, state):
    """Genera il codice per un singolo movimento."""
    move_code = f"(mv_{time}_{moving_tile}_{current_pos}_{new_pos} -> ("

    # update pos of moving tile 
    move_conditions = [
        f"tile{moving_tile}_pos{new_pos}_{time + 1}",
        f"!tile{moving_tile}_pos{current_pos}_{time + 1}"
    ]

    # Apdate empty
    move_conditions += [
        f"empty_pos{current_pos}_{time + 1}",
        f"!empty_pos{new_pos}_{time + 1}"
    ]

    # Add condition for tile that do not move
    for i in range(len(state)):
        for j in range(len(state[i])):
            tile = state[i][j]
            pos = i * len(state) + j + 1
            if tile != 0 and pos != current_pos and pos != new_pos:
                move_conditions.append(f"tile{tile}_pos{pos}_{time + 1}")

    move_code += " & ".join(move_conditions) + ")) &\n"
    return move_code

def find_adjacent_tiles(state, empty_pos):
    """Trovare i blocchi adiacenti alla posizione vuota."""
    adjacent_tiles = []
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # sopra, sotto, sinistra, destra

    for di, dj in directions:
        new_i, new_j = empty_pos[0] + di, empty_pos[1] + dj
        if 0 <= new_i < len(state) and 0 <= new_j < len(state[0]):
            adjacent_tiles.append((state[new_i][new_j], new_i * len(state) + new_j))

    return adjacent_tiles

def generate_movement_code(time, tile, tile_pos, empty_pos):
    """Genera il codice per un singolo movimento."""
    move_code = f"(mv_{time}_{tile}_{tile_pos}_{empty_pos} -> ("
    move_conditions = [
        f"tile{tile}_pos{empty_pos + 1}_{time + 1}", 
        f"!tile{tile}_pos{tile_pos + 1}_{time + 1}", 
        f"empty_pos{tile_pos + 1}_{time + 1}",
        f"!empty_pos{empty_pos + 1}_{time + 1}"
    ]
    move_code += " & ".join(move_conditions) + ")) &\n"
    return move_code

def generate_possible_moves(state, time):
    """Genera tutti i possibili movimenti a partire dallo stato corrente."""
    empty_pos = find_empty_tile(state)
    empty_pos_index = empty_pos[0] * len(state[0]) + empty_pos[1] + 1  # 
    moves = []

    # Possible movements
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for di, dj in directions:
        new_i, new_j = empty_pos[0] + di, empty_pos[1] + dj

        if 0 <= new_i < len(state) and 0 <= new_j < len(state[0]):
            moving_tile = state[new_i][new_j]
            if moving_tile != 0:  # no space empty
                current_pos_index = new_i * len(state[0]) + new_j + 1 
                move_code = create_movement_code(time, moving_tile, current_pos_index, empty_pos_index, state)
                moves.append(move_code)

    return moves

def generate_movement_constraints(all_moves):
    """Genera i vincoli di movimento per un numero variabile di mosse."""
    constraints = "% Movement constraints\n"
    num_times = len(all_moves)

    # Se abbiamo solo un momento temporale, non ci sono vincoli da aggiungere
    if num_times <= 1:
        return ""

    # Per ogni momento temporale, tranne l'ultimo, genera i vincoli
    for time in range(num_times - 1):
        for move in all_moves[time]:
            move_code = move.split(' -> ')[0][1:]
            constraints += f"(({move_code}) -> ("

            # Aggiungi i codici dei movimenti possibili al momento successivo come alternativa
            for next_move in all_moves[time + 1]:
                next_move_code = next_move.split(' -> ')[0][1:]
                constraints += f"{next_move_code} | "

            constraints = constraints[:-3] + ")) &\n"  # Rimuove l'ultimo ' | ' e chiude la parentesi

    return constraints


def generate_frame_axioms(all_moves):
    """Genera gli assiomi di frame basati sui movimenti per un numero variabile di mosse."""
    axioms = "% Frame axioms\n"
    
    for time_moves in all_moves:
        for move in time_moves:
            move_code = move.split(' -> ')[0][1:]
            move_conditions = move.split(' -> ')[1]
            axioms += f"(({move_code}) -> {move_conditions}\n"

    return axioms

def generate_sat_code(initial_state, num_moves):
    """Genera il codice SAT completo per il puzzle."""
    sat_code = "% Initial State\n"
    for i in range(len(initial_state)):
        for j in range(len(initial_state[i])):
            tile = initial_state[i][j]
            if tile != 0:
                sat_code += f"tile{tile}_pos{i * len(initial_state) + j + 1}_0 &\n"
            else:
                sat_code += f"empty_pos{i * len(initial_state) + j + 1}_0 &\n"

    sat_code += "% Final State\n"
    for i in range(1, len(initial_state)**2 + 1):
        sat_code += f"tile{i}_pos{i}_{num_moves} &\n" if i != len(initial_state)**2 else f"empty_pos{i}_{num_moves} &\n"

    # Genera i movimenti per ogni momento temporale
    all_moves = [generate_possible_moves(initial_state, 0)]
    for time in range(1, num_moves):
        new_moves = []
        for move in all_moves[-1]:
            new_state = apply_move(initial_state, move)
            new_moves.extend(generate_possible_moves(new_state, time))
        all_moves.append(new_moves)

    # Aggiungi i movimenti al codice SAT
    sat_code += "% Movements\n"
    for moves in all_moves:
        for move in moves:
            sat_code += move

    # Genera le stringhe di vincoli e assiomi
    movement_constraints = generate_movement_constraints(all_moves)
    frame_axioms = generate_frame_axioms(all_moves)

    # Aggiungi le stringhe generate al codice SAT
    sat_code += movement_constraints
    sat_code += frame_axioms
    return sat_code.rstrip(" &\n")

# Apply movement of the space
def apply_move(state, move):
    """Applica un movimento allo stato e restituisce il nuovo stato."""
    move_details = move.split(' -> ')[0]  # take the part before of '->'
    _, moving_tile, current_pos, new_pos = [int(x) for x in move_details.split('_')[1:]]

    current_i, current_j = (current_pos - 1) // len(state), (current_pos - 1) % len(state)
    new_i, new_j = (new_pos - 1) // len(state), (new_pos - 1) % len(state)

    # Create a copy of the state and apply the movement
    new_state = [row[:] for row in state]
    new_state[current_i][current_j], new_state[new_i][new_j] = new_state[new_i][new_j], new_state[current_i][current_j]
    return new_state

# Test
initial_state = [[1, 2, 3], [8, 7, 4], [6, 5, 0]]  #  puzzle 3x3
sat_code = generate_sat_code(initial_state, 12)  
print(sat_code)

file_path = "project1.txt"

with open(file_path, 'w') as file:
    # Write the string to the file
    file.write(sat_code)

result = run_limboole_with_exe(sat_code)
print(result)
```
