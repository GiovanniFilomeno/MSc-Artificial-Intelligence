```python
import matplotlib.pyplot as plt
import numpy as np # Optional, but good practice for numerical data

# Data for the 9-day biking trip
distances = [45, 50, 48, 52, 47, 49, 51, 46, 58]

# Create the figure and axes for the plot
fig, ax = plt.subplots(figsize=(6, 4)) # Adjust figsize as needed

# Create the boxplot
# 'vert=False' makes the boxplot horizontal
# 'patch_artist=True' fills the box with color
# 'showfliers=True' explicitly shows outliers if any (though none here)
boxplot_dict = ax.boxplot(distances, vert=False, patch_artist=True, showfliers=True)

# --- Optional Customizations ---
# Set colors (example: light blue box, black whiskers/median)
colors = ['lightblue']
for patch in boxplot_dict['boxes']:
    patch.set_facecolor(colors[0])

# Set whisker and cap color
for whisker in boxplot_dict['whiskers']:
    whisker.set(color='black', linewidth=1.5)
for cap in boxplot_dict['caps']:
    cap.set(color='black', linewidth=1.5)

# Set median line color
for median in boxplot_dict['medians']:
    median.set(color='black', linewidth=1.5)

# --- Labels and Title ---
ax.set_xlabel("Distance (km)")
ax.set_yticks([]) # Hide y-axis ticks as it's just one boxplot
ax.set_title("Daily Distances for 9-Day Biking Trip")

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

midpoints = [1050, 1150, 1250, 1350, 1450]
freq =      [   16,   18,   15,   11,    0]

data_expanded = []
for m, f in zip(midpoints,freq):
    data_expanded.extend([m]*f)

plt.boxplot(data_expanded)
plt.title("Rabbit Weights (grams)")
plt.ylabel("Weight (g)")
plt.show()
```
